```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
times = list(np.arange(0, 4.1, 0.5))
times_2s = [0, 1, 2, 3]
times_3s = [0, 1, 2]
times_4s = [0, 1]
```

```python
rh_1s = [85.11, 82.98, 86.44, 82.23, 87.51, 85.64, 87.4, 88.07, 93.5]
rh_2s = [86.91, 86.92, 87.38, 88.12]
rh_3s = [88.0, 91.78, 93.29]
rh_4s = [89.94, 93.47]
```

```python
plt.ylim(70, 100)
plt.plot(times, rh_1s, marker='o')
plt.plot(times_2s, rh_2s, marker='o')
plt.plot(times_3s, rh_3s, marker='o')
plt.plot(times_4s, rh_4s, marker='o')
plt.axhline(y=96, label='5 second window', alpha=0.6, linestyle='--')
plt.xlabel('Start of time window (seconds)')
plt.ylabel('Validation accuracy (%)')
plt.legend(['1 second window', '2 second window', '3 second window', '4 second window', '5 second window'])
plt.grid()
plt.title('Sliding window SVM classification results for PP/PNP-RH')
plt.savefig('slid_rh.jpg')
```

```python
l_1s = [72.33, 80.09, 83.25, 80.35, 69.8, 54.55, 55.69, 59.47, 43.2]
l_2s = [89.77, 86.86, 81.75, 70.73]
l_3s = [90.78, 92.19, 76.5]
l_4s = [93.9, 94.06]
```

```python
plt.ylim(40, 100)
plt.plot(times, l_1s, marker='o')
plt.plot(times_2s, l_2s, marker='o')
plt.plot(times_3s, l_3s, marker='o')
plt.plot(times_4s, l_4s, marker='o')
#plt.plot([0], [94], marker='r--')
plt.axhline(y=94, label='5 second window', alpha=0.6, linestyle='--')
plt.xlabel('Start of time window (seconds)')
plt.ylabel('Validation accuracy (%)')
plt.legend(['1 second window', '2 second window', '3 second window', '4 second window', '5 second window'])
plt.grid()
plt.title('Sliding window SVM classification results for PP/PNP-L')
plt.savefig('slid_l.jpg')
```
